```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=pd.read_csv("data.csv")
```

```python
plt.figure(figsize=(12, 6)) 

sns.barplot(x='State', y='Population', data=x)

plt.xlabel('State')
plt.ylabel('Population')
plt.title('Population by State')

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6)) 

sns.barplot(x='State', y='Males', data=x)

plt.xlabel('State')
plt.ylabel('Male Population')
plt.title('Male Population by State')

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6)) 

sns.barplot(x='State', y='Females', data=x)

plt.xlabel('State')
plt.ylabel('Female Population')
plt.title('Female Population by State')


plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
